```python
from bs4 import BeautifulSoup as bs
```

```python
def get_bullets(mysoup):
    rs=mysoup.find_all("div", class_="_1n-_ _6hy- _8lf-")

    bullets = []
    for x in rs:
        bullets.append(str(x.text))
    
    return ('['+']<br/>['.join(bullets)+']')
```

```python
def get_title(mysoup):
    return (mysoup.find_all("div", class_="_1n-z _6hy- _8lfs")[0].text)  
```

```python
def get_description(mysoup):
    return (mysoup.find_all("div", class_="_1n-_ _6hy- _94t2")[0].text)
```

```python
def get_locations(mysoup):
    rs=mysoup.find_all("div", class_="_8mle")
        
    locs = []
    for x in rs:
        locs.append(str(x.text))
    
    return (' |<br/> '.join(locs))
```

```python
def get_fb_link(job_id):
    return ("https://www.facebook.com/careers/jobs/"+job_id)

def make_clickable(val):
    jid=val.split("/")[-1:][0]
    return f'<a href="{get_fb_link(val)}" target="_blank">{jid}</a>'
```

```python
df = pd.DataFrame(columns=['job_id', 
                           'title', 
                           'location', 
                           'bullets', 
                           'description'])

i = 0 
with os.scandir("../data/") as my_dir:
    for entry in my_dir:
        #print(entry.name, entry.is_file ())
        with open(entry, 'r') as myfile:
            html_contents = myfile.read()
            soup = bs(html_contents, 'lxml')   
            df.loc[i] = [entry.name, 
                         get_title(soup), 
                         get_locations(soup),
                         get_bullets(soup), 
                         get_description(soup)]
            i = i + 1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.feature_extraction import text 

my_additional_stop_words = frozenset([
    "br", "closely", "degree", "educational",
    "equivalent", "experience", "knowledge", 
    "languages", "number",
    "product", 
    "solutions", "underlying",
    "using", "variety",
    "work", "working",
    "years"
])

my_stop_words = text.ENGLISH_STOP_WORDS.union(my_additional_stop_words)
```

```python
vectorizer = CountVectorizer(stop_words=my_stop_words)
```

```python
dataset_size = len(df.bullets)
print(f"The dataset contains {dataset_size} job posts")
```

```python
tf_matrix = vectorizer.fit_transform(df.bullets)
#print(tf_matrix)
```

```python
tf_np_matrix = tf_matrix.toarray()
#print(tf_np_matrix)
```

```python
assert tf_np_matrix.shape == tf_matrix.shape
num_posts, vocabulary_size = tf_np_matrix.shape
print(f"The collection of {num_posts} job posts contain a total of "
      f"{vocabulary_size} unique words")
```

```python
import numpy as np
tf_vector = tf_np_matrix[0]
non_zero_indices = np.flatnonzero(tf_vector)
num_unique_words = non_zero_indices.size
print(f"The job in row 0 contains {num_unique_words} unique words.")
#print("The actual word-counts map to the following column indices:\n")
#print(non_zero_indices)
```

```python
words = vectorizer.get_feature_names()
unique_words = [words[i] for i in non_zero_indices]
print(unique_words)
```

```python
import pandas as pd
data = {'Word': unique_words,
        'Count': tf_vector[non_zero_indices]}

df2 = pd.DataFrame(data).sort_values('Count', ascending=False)
print(df2[:10].to_string(index=False))
```

```python
sub_matrix = tf_np_matrix[:,non_zero_indices]
#print("We obtained a sub-matrix correspond to the 34 words within post 0. "
#      "The first row of the sub-matrix is:")
print(sub_matrix[0])
```

```python
from sklearn.preprocessing import binarize
binary_matrix = binarize(sub_matrix)
#print(binary_matrix)
```

```python
unique_post_mentions = binary_matrix.sum(axis=0)
#print("This vector counts the unique posts in which each word is "
#      f"mentioned:\n {unique_post_mentions}")
```

```python
np_post_mentions = binarize(tf_np_matrix[:,non_zero_indices]).sum(axis=0)
csr_post_mentions = binarize(tf_matrix[:,non_zero_indices]).sum(axis=0)
#print(f'NumPy matrix-generated counts:\n {np_post_mentions}\n')
#print(f'CSR matrix-generated counts:\n {csr_post_mentions}')
```

```python
document_frequencies = unique_post_mentions / dataset_size
data3 = {'Word': unique_words,
        'Count': tf_vector[non_zero_indices],
        'Document Frequency': document_frequencies}

df3 = pd.DataFrame(data3)
df3_common_words = df3[df3['Document Frequency'] >= .1]
print(df3_common_words.to_string(index=False))
```

```python
df3_sorted = df3.sort_values(['Count', 'Document Frequency'],
                           ascending=[False, True])
print(df3_sorted[:10].to_string(index=False))
```

```python
inverse_document_frequencies = 1 / document_frequencies
df3['IDF'] = inverse_document_frequencies
df3['Combined'] = df3.Count * inverse_document_frequencies
df3_sorted = df3.sort_values('Combined', ascending=False)
print(df3_sorted[:10].to_string(index=False))
```

```python
df3['Combined'] = df3.Count * np.log10(df3.IDF)
df3_sorted = df3.sort_values('Combined', ascending=False)
print(df3_sorted[:10].to_string(index=False))
```

```python
keywords = ["java", "python"]

lst = list()

for k in keywords:
    try:
        print("{} - {}".format(k,words.index(k)))
        lst.append(words.index(k))
    except ValueError:
        print(f"{k} not in list.")
```

```python
keywords_non_zero_indices =  np.ndarray((len(lst),), buffer=np.array(lst),
           dtype=int) 
keywords_non_zero_indices
```

```python
tf_np_matrix.shape
```

```python
sub_matrix = tf_np_matrix[:,keywords_non_zero_indices]
sub_matrix
```

```python
binary_matrix = binarize(sub_matrix)
```

```python
posts_with_at_least_one = binary_matrix.sum(axis=1)
posts_with_at_least_one = (posts_with_at_least_one >= 1).astype(np.int_)
posts_with_at_least_one
```

```python
posts_of_interest = np.flatnonzero(posts_with_at_least_one)
```

```python
df_of_interest = pd.DataFrame(columns=['job_id', 
                           'title', 
                           'location', 
                           'bullets', 
                           'description'])

i = 0 

for p in posts_of_interest:
    df_of_interest.loc[i] = df.loc[p]
    i = i + 1
```
